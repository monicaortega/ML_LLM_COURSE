# Cadenas con Langchain


Una cadena en LangChain es un flujo o secuencia de operaciones que permite encadenar múltiples pasos de procesamiento de lenguaje, típicamente incluyendo:

- Entrada de datos
- Transformación de prompts
- Interacción con modelos de lenguaje
- Procesamiento de respuestas

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import PromptTemplate
```

```python
OPENAI_API_KEY = ""
```

```python
llm = ChatOpenAI(
    model='gpt-3.5-turbo',
    temperature=0.7,  # Creativity level
    max_tokens=250,   # Maximum response length
    frequency_penalty=0.5,  # Reduce repetition
    presence_penalty=0.5,   # Encourage novel topics
    openai_api_key=OPENAI_API_KEY
)
```

### Cadena simple

```python


template = """
    Role: {persona}
    Task: {instruction}
    Context: {additional_context}
    """
prompt = PromptTemplate(
    input_variables=['persona', 'instruction', 'additional_context'],
    template=template
)

chain = (
    {"persona": RunnablePassthrough(), 
        "instruction": RunnablePassthrough(), 
        "additional_context": RunnablePassthrough()} 
    | prompt 
    | llm 
    | StrOutputParser()
)

response = chain.invoke({
    'persona': 'Technology Expert',
    'instruction': 'Tell me something about dolphins',
    'additional_context': 'For 5 years old kids'
})
print(response)
```

**PromptTemplate**
<p>Los PromptTemplates en LangChain son fundamentales por varias razones:<p>

- Parametrización Dinámica: Permiten crear prompts flexibles que pueden cambiar según la entrada.

- Separación de Concerns: Desacoplan la estructura del prompt de los datos específicos.

- Reutilización: Facilitan la creación de prompts consistentes y reutilizables.

- Control de Variables: Aseguran que solo se usen las variables definidas.

```python
cliente = OpenAI(api_key=OPENAI_API_KEY)
# Método 1: Sin PromptTemplate (Manual)
def generar_idea_manual(categoria, problema):
    prompt_manual = f"""
    Genera una idea innovadora para {categoria}
    resolviendo: {problema}
    """

    # Hacer una llamada al modelo
    respuesta = cliente.chat.completions.create(
        model="gpt-3.5-turbo",  # Puedes cambiar a gpt-4 si tienes acceso
        messages=[
            {"role": "user", "content": prompt_manual}
        ],
        temperature=0.7,  # Controla la creatividad (0 = determinístico, 1 = más creativo)
        max_tokens=150    # Limita la longitud de la respuesta
    )
    return respuesta.choices[0].message.content
generar_idea_manual("productividad","gestion de tareas")
```

#### Few-shot template

```python
from langchain_core.prompts import FewShotPromptTemplate

examples = [
    {"input": "feliz", "output": "positivo"},
    {"input": "triste", "output": "negativo"}
]

# Template con ejemplos
few_shot_prompt = FewShotPromptTemplate(
    example_prompt=PromptTemplate(
        input_variables=["input", "output"],
        template="Entrada: {input}, Salida: {output}"
    ),
    examples=examples,
    input_variables=["input"],
    prefix="Clasifica una palabra basandote en los siguientes ejemplos", 
    suffix="Entrada: {input}\nSalida:", 
)
```

```python
chain_few_shot = (
        {"input": RunnablePassthrough()} 
        | few_shot_prompt 
        | llm 
        | StrOutputParser()
    )
```

```python
chain_few_shot.invoke({"input": "unicornio"})
```

##### Prefix y suffix
**Prefix**

- Va al INICIO del prompt
- Proporciona contexto general
- Establece el marco de trabajo
- Ayuda al modelo a entender el objetivo

✅ Usar Prefix cuando necesites:

- Explicar el contexto general
- Dar instrucciones iniciales
- Establecer el tono o estilo

**Suffix**

- Va al FINAL del prompt
- Define el formato de salida
- Completa las instrucciones
- Indica cómo procesar los ejemplos previos


✅ Usar Suffix cuando necesites:

- Definir el formato de respuesta
- Indicar cómo procesar los ejemplos
- Completar las instrucciones de la tarea
