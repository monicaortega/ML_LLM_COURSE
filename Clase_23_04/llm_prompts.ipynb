### Consejos para realizar un buen prompt (Prompt Engineering)

Mejores Prácticas para la Ingeniería de Prompts
1. **Instrucciones Claras**
- Sé específico y explícito
- Una tarea por prompt
- Usa palabras de acción
Ejemplo:
❌ "Háblame sobre coches"
✅ "Explica los componentes básicos de un motor de gasolina moderno y sus funciones"

2. **Configura un contexto**

- Proporciona información relevante
- Define el ámbito/tema
- Especifica la audiencia objetivo
Ejemplo:
❌ "Escribe sobre el cambio climático"
✅ "Escribe una explicación sobre el cambio climático para estudiantes de secundaria, enfocándote en conceptos básicos y ejemplos cotidianos"

3. **Especifica el formato**

- Define la estructura de salida deseada
- Especifica la cantidad de elementos
- Solicita el tipo de ejemplos, si es necesario
Ejemplo:
❌ "Enumera alimentos saludables"
✅ "Crea una lista de 5 alimentos saludables, incluye para cada uno: nombre, nutrientes clave y una receta simple. Formatea cada entrada con viñetas."

4. **Few-Shot Learning**
- Añade ejemplos del output deseado
- Muestra el patrón de interacción que deseas
Ejemplo:
Convierte las siguientes frases a pasado:
Input: "Yo voy a la tienda"
Output: "Yo fui a la tienda"
Input: "Ella corre rápido"
Output: "Ella corría rápido"
Ahora convierte: "Ellos cantan bien"

5. **Cadena de pensamiento (Chain of Thought)**

- Divide las tareas complejas en tareas en pasos
- Muestra el proceso de razonamiento
Por ejemplo:

Pregunta: ¿Cuánto es el 15% de 80?
Vamos a resolverlo paso a paso:
1. Convierte el 15% a decimal (15/100 =0.15)
2. Multiplica 80 por 0.15
3. 80 x 0.15 =12
Por tanto, el 15% de 80 es 12

6. **Configura el rol y la perspectiva**
- Define el personaje de la IA
- Especifica la perspectiva de la explicación


#### Errores comunes a evitar
**Ambigüedad**
❌ "Hazlo mejor"
✅ "Mejora el texto haciendo las oraciones más cortas y usando palabras más simples"

**Excesiva complejidad**
❌ "Explica la física cuántica, la relatividad y la teoría de cuerdas"
✅ "Explica un concepto básico de la física cuántica: la superposición"

**Falta de Restricciones**
❌ "Escribe una historia"
✅ "Escribe una historia de 200 palabras sobre la amistad, adecuada para niños de 8-10 años"

### OpenAI completion parameters
El método de Completion.create se utiliza para tareas generales de generación de texto. Los parámetros clave incluyen:
- model: El ID del modelo a utilizar (por ejemplo, "text-davinci-003")
- prompt: El texto de entrada para generar las completaciones
- max_tokens: El número máximo de tokens a generar (Defecto=16)
- temperature: Controla la aleatoriedad (0-1, valores más altos significan más aleatoriedad)(Defecto=1)
- presence_penalty: Penaliza nuevos tokens basándose en su presencia en el texto hasta ahora([-2,2]) (Defecto=0)
- frequency_penalty: Penaliza nuevos tokens basándose en su frecuencia en el texto hasta ahora([-2,2]) (Defecto=0)

+info: https://platform.openai.com/docs/api-reference/completions/create

### Role-Playing Prompts

```python
prompts = [
    "Explica la IA",
    "Explica la IA como si fueras un profesor de secundaria",
    "Explica la IA como si fueras el CEO de una startup tecnológica"
]

for p in prompts:
    respuesta = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": f"{p}",
        }
    ],
    model="gpt-3.5-turbo"
)
    print(f"Prompt: {p}\nResponse: {respuesta.choices[0].message.content}\n---")

```

### Zero-shot vs. Few-shot learning

```python
#Zero-shot
prompt = "Escribe una descripción de producto para un reloj inteligente:"

respuesta = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": f"{prompt}",
        }
    ],
    model="gpt-4"
)

respuesta.choices[0].message.content
```

```python
#Few-shot
few_shot_examples = """
Ejemplo 1: Fitness tracker que monitorea la frecuencia cardíaca, los pasos y la calidad del sueño.
Perfecto para personas conscientes de su salud que buscan información integral sobre bienestar.

Ejemplo 2: Auriculares inalámbricos con cancelación de ruido, 24 horas de batería y audio cristalino.
Ideal para amantes de la música y profesionales que exigen una experiencia de sonido premium.
"""

prompt = f"""{few_shot_examples}
Escribe en castellano una descripción de producto para un reloj inteligente:"""

respuesta = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": f"{prompt}",
        }
    ],
    model="gpt-3.5-turbo"
)
respuesta.choices[0].message.content
```

### Ejercicio "de NLTK a las LLMS"

Las entidades y las intenciones (entities and intents) son un concepto básico del Entendimiento del Lenguaje Natural (NLU)

**Entidades** El Reconocimiento de Entidades Nombradas (NER, por sus siglas en inglés) identifica piezas específicas de información en un texto, como nombres, fechas, ubicaciones y organizaciones.

Ejemplos:
"Barack Obama" → Persona
"New York" → Ubicación

**Intenciones** Es la comprensión del propósito o intención detrás de la entrada de un usuario. A menudo se enmarca como clasificación de intenciones en NLP.

Ejemplos:
"Reservar un vuelo a París." → Intención: Reservar
"¿Cuál es el clima en Londres?" → Intención: Consulta sobre el clima

#### Reconocimiento de entidades

```python
from nltk import word_tokenize, pos_tag, ne_chunk
```

```python
# Download required resources
nltk.download('punkt', download_dir=nltk_data_path)
nltk.download('maxent_ne_chunker', download_dir=nltk_data_path)
nltk.download('words', download_dir=nltk_data_path)
nltk.download('averaged_perceptron_tagger', download_dir=nltk_data_path)
nltk.download('averaged_perceptron_tagger_eng', download_dir=nltk_data_path)
nltk.download('punkt_tab', download_dir=nltk_data_path)
nltk.download('maxent_ne_chunker_tab')
```

```python

# Example sentence
sentence = "Barack Obama nació en Hawaii y fue el presidente de los Estados Unidos."

# Tokenize, Part-of-Speech Tagging, and Named Entity Recognition
tokens = word_tokenize(sentence)
pos_tags = pos_tag(tokens)
entities = ne_chunk(pos_tags)

# Display entities
print("Entidades:", entities)

entities
```

```python
from nltk.tree import Tree
def extract_entities(tree):
    entities = []
    for subtree in tree:
        if isinstance(subtree, Tree):
            entity = " ".join([word for word, tag in subtree.leaves()])
            label = subtree.label()
            entities.append((entity, label))
    return entities
```

```python
extract_entities(entities)
```

#### Intención del usuario

```python
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import names
```

```python
# Example training data
training_data = [
    ("Reserva un viaje a Nueva York", "Reserva"),
    ("Cancela mi reserva", "Reserva"),
    ("¿Qué tiempo hace en Londres?", "Tiempo"),
    ("¿Lloverá mañana?", "Tiempo"),
]

# Prepare data for training
def format_data(data):
    return [({"text": tuple(sentence.split())}, label) for sentence, label in data]

training_set = format_data(training_data)

# Train a simple classifier
classifier = NaiveBayesClassifier.train(training_set)

# Test the classifier
test_sentence = "Quiero comprarme unos zapatillas"
print("Intención:", classifier.classify({"text": tuple(test_sentence.split())}))
```
