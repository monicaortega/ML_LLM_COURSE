# Tokenización y Embeddings para Principiantes

## Introducción

En este notebook, aprenderemos sobre dos conceptos fundamentales para trabajar con Modelos de Lenguaje (LLM):

1. **Tokenización**: El proceso de dividir texto en unidades más pequeñas llamadas "tokens"
2. **Embeddings**: Representaciones numéricas de palabras o frases en un espacio vectorial

Estos conceptos son la base para que los modelos de lenguaje puedan procesar y entender el texto.

## Parte 1: Tokenización

La tokenización es el primer paso para procesar texto. Consiste en dividir el texto en unidades más pequeñas (tokens) que los modelos pueden entender.

### 1.1 Tokenización básica con Python

```python
# Método simple: dividir por espacios
texto = "Hola, ¿cómo estás aprendiendo sobre LLMs?"
tokens_simple = texto.split()
print(f"Tokens simples: {tokens_simple}")
print(f"Número de tokens: {len(tokens_simple)}")

# Un poco más elaborado: eliminando signos de puntuación
import re
tokens_sin_puntuacion = re.findall(r'\b\w+\b', texto.lower())
print(f"Tokens sin puntuación: {tokens_sin_puntuacion}")
print(f"Número de tokens: {len(tokens_sin_puntuacion)}")
```

### 1.2 Tokenización con Tiktoken (usado por OpenAI)

Tiktoken es la biblioteca que usa OpenAI para tokenizar texto para sus modelos como GPT.

```python
import tiktoken

# Cargamos el codificador para el modelo GPT-3.5-turbo
codificador = tiktoken.get_encoding("cl100k_base")  # Codificador usado por gpt-3.5-turbo y gpt-4

texto = "Hola, ¿cómo estás aprendiendo sobre LLMs?"
tokens = codificador.encode(texto)

print(f"Tokens (IDs): {tokens}")
print(f"Número de tokens: {len(tokens)}")

# Decodificar para ver cómo se divide el texto
for token in tokens:
    print(f"Token ID: {token}, Token: {codificador.decode([token])}")
```

## Parte 2: Embeddings

Los embeddings son representaciones vectoriales densas de palabras o frases. Permiten capturar el significado semántico y contextual del texto.

```python
OPENAI_API_KEY = ""
```

### 2.1 Embeddings con OpenAI

```python
! pip install openai
```

```python
from openai import OpenAI
import os

# Nota: Necesitas configurar tu API key de OpenAI
# os.environ["OPENAI_API_KEY"] = "tu-api-key"

# Intentar usar la API key si está configurada
try:
    cliente = OpenAI(api_key=OPENAI_API_KEY)
    
    texto = "Los embeddings son representaciones vectoriales del significado de un texto."
    
    respuesta = cliente.embeddings.create(
        input=texto,
        model="text-embedding-3-small"
    )
    
    embedding = respuesta.data[0].embedding
    
    print(f"Dimensión del embedding: {len(embedding)}")
    print(f"Primeros 5 valores del embedding: {embedding[:5]}")
    print(f"El embedding completo es un vector de {len(embedding)} dimensiones.")
    
except Exception as e:
    print(f"No se pudo generar el embedding con OpenAI: {e}")
    print("Para usar esta funcionalidad, necesitas configurar tu API key.")
```

### 2.2 Comparación de Embeddings con OpenAI

```python
# Función para calcular similitud de coseno entre dos vectores
import numpy as np

def similitud_coseno(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)

# Intentar usar la API de OpenAI si está configurada
try:
    cliente = OpenAI(api_key=OPENAI_API_KEY)
    
    texto1 = "Los gatos son mascotas populares."
    texto2 = "Los felinos domésticos son queridos como animales de compañía."
    texto3 = "Los algoritmos de machine learning requieren muchos datos."
    
    # Obtener embeddings
    respuesta1 = cliente.embeddings.create(input=texto1, model="text-embedding-3-small")
    respuesta2 = cliente.embeddings.create(input=texto2, model="text-embedding-3-small")
    respuesta3 = cliente.embeddings.create(input=texto3, model="text-embedding-3-small")
    
    embedding1 = respuesta1.data[0].embedding
    embedding2 = respuesta2.data[0].embedding
    embedding3 = respuesta3.data[0].embedding
    
    # Calcular similitud
    sim_1_2 = similitud_coseno(embedding1, embedding2)
    sim_1_3 = similitud_coseno(embedding1, embedding3)
    sim_2_3 = similitud_coseno(embedding2, embedding3)
    
    print(f"Similitud entre texto1 y texto2: {sim_1_2:.4f}")  # Debería ser alta
    print(f"Similitud entre texto1 y texto3: {sim_1_3:.4f}")  # Debería ser baja
    print(f"Similitud entre texto2 y texto3: {sim_2_3:.4f}")  # Debería ser baja
    
except Exception as e:
    print(f"No se pudo hacer la comparación de embeddings: {e}")
    print("Para usar esta funcionalidad, necesitas configurar tu API key.")
```

## Parte 3: Introducción a LangChain y llamadas a modelos GPT

Antes de pasar a las aplicaciones prácticas, vamos a ver cómo hacer llamadas a modelos de lenguaje (LLMs) como GPT utilizando dos enfoques:

1. Usando directamente la API de OpenAI
2. Usando LangChain como framework intermediario

### ¿Qué es LangChain?

LangChain es un framework diseñado para facilitar el desarrollo de aplicaciones potenciadas por modelos de lenguaje (LLMs). Ofrece varias ventajas:

- **Abstracciones consistentes**: Proporciona interfaces estandarizadas para diferentes proveedores de LLMs (OpenAI, Anthropic, Cohere, etc.)
- **Componentes reutilizables**: Ofrece herramientas para tareas comunes como recuperación de documentos, cadenas de razonamiento, agentes, etc.
- **Composición de flujos**: Permite construir aplicaciones complejas al combinar diferentes componentes
- **Persistencia y rastreo**: Facilita el seguimiento de las interacciones con LLMs

En resumen, LangChain actúa como una capa intermedia que simplifica la interacción con LLMs y permite crear aplicaciones más sofisticadas sin tener que reinventar soluciones comunes.

### 2.4.1 Llamadas directas a GPT con OpenAI

Primero, veamos cómo hacer llamadas directas al API de OpenAI:

```python
from openai import OpenAI
import os

# Configura tu API key (recomendable hacerlo con variables de entorno)
# os.environ["OPENAI_API_KEY"] = "tu-api-key"

try:
    # Inicializar el cliente
    cliente = OpenAI(api_key=OPENAI_API_KEY)
    
    # Definir un prompt simple
    prompt = "Explica en dos frases qué es la tokenización en el contexto de PLN."
    
    # Hacer una llamada al modelo
    respuesta = cliente.chat.completions.create(
        model="gpt-3.5-turbo",  # Puedes cambiar a gpt-4 si tienes acceso
        messages=[
            {"role": "system", "content": "Eres un asistente útil especializado en procesamiento de lenguaje natural."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7,  # Controla la creatividad (0 = determinístico, 1 = más creativo)
        max_tokens=150    # Limita la longitud de la respuesta
    )
    
    # Mostrar la respuesta
    print("Respuesta de GPT:\n")
    print(respuesta.choices[0].message.content)
    
    # Mostrar información sobre el uso de tokens
    print("\nInformación de tokens:")
    print(f"Tokens de entrada: {respuesta.usage.prompt_tokens}")
    print(f"Tokens de salida: {respuesta.usage.completion_tokens}")
    print(f"Total tokens: {respuesta.usage.total_tokens}")
    
except Exception as e:
    print(f"Error al hacer la llamada a OpenAI: {e}")
    print("Asegúrate de haber configurado correctamente tu API key.")
```

### 2.4.3 Llamadas a GPT con LangChain

Ahora, veamos cómo hacer lo mismo pero utilizando LangChain como intermediario:

```python
! pip install langchain_openai
! pip install langchain_core
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
# Initialize LLM
llm = ChatOpenAI(
    model='gpt-3.5-turbo',
    temperature=0.7,  # Creativity level
    max_tokens=250,   # Maximum response length
    frequency_penalty=0.5,  # Reduce repetition
    presence_penalty=0.5,   # Encourage novel topics
    openai_api_key=OPENAI_API_KEY
)
```

```python
# 1. Basic Prompt Template
def basic_prompt():
    template = """
    Role: {persona}
    Task: {instruction}
    Context: {additional_context}
    """
    prompt = PromptTemplate(
        input_variables=['persona', 'instruction', 'additional_context'],
        template=template
    )
    
    chain = (
        {"persona": RunnablePassthrough(), 
         "instruction": RunnablePassthrough(), 
         "additional_context": RunnablePassthrough()} 
        | prompt 
        | llm 
        | StrOutputParser()
    )
    
    response = chain.invoke({
        'persona': 'Technology Expert',
        'instruction': 'Tell me something about dolphins',
        'additional_context': 'For 5 years old kids'
    })
    return response

respuesta=basic_prompt()
```

### 2.4.6 Comparación: API de OpenAI vs LangChain

| Aspecto | API de OpenAI directa | LangChain |
|---------|----------------------|----------|
| **Simplicidad** | Más directo para casos de uso simples | Requiere aprender abstracciones adicionales |
| **Flexibilidad** | Limitado a las funciones de la API | Alta flexibilidad con componentes modulares |
| **Composición** | Requiere código personalizado para flujos complejos | Facilita la composición de componentes |
| **Múltiples modelos** | Limitado a OpenAI | Soporta múltiples proveedores de LLMs |
| **Persistencia** | Requiere implementación manual | Incluye herramientas para rastreo y persistencia |
| **Curva de aprendizaje** | Baja | Media-Alta |

**¿Cuándo usar cada uno?**

- **API de OpenAI directa**: Para proyectos pequeños, prototipos rápidos o cuando solo necesitas hacer llamadas simples a GPT.
- **LangChain**: Para aplicaciones más complejas, cuando necesitas integrar múltiples componentes, trabajar con varios proveedores de modelos, o quieres aprovechar herramientas existentes para tareas comunes.

En general, LangChain brinda mayor escalabilidad y reutilización, mientras que la API directa ofrece más simplicidad para casos de uso básicos.

### 2.3 Embeddings con LangChain

LangChain proporciona una interfaz unificada para trabajar con varios proveedores de embeddings.

```python
from langchain_openai import OpenAIEmbeddings
from langchain_core.embeddings import Embeddings

# Intentar usar embeddings de OpenAI a través de LangChain
try:
    # Inicializar el modelo de embeddings
    embeddings = OpenAIEmbeddings(model="text-embedding-3-small",api_key=OPENAI_API_KEY)
    
    # Generar embeddings para un texto
    texto = "LangChain facilita el trabajo con LLMs."
    vector = embeddings.embed_query(texto)
    
    print(f"Dimensión del embedding: {len(vector)}")
    print(f"Primeros 5 valores: {vector[:5]}")
    
    # Generar embeddings para varios textos
    textos = [
        "Los embeddings son útiles para búsqueda semántica.",
        "La similitud de coseno mide la cercanía de vectores.",
        "Los modelos de lenguaje utilizan embeddings internamente."
    ]
    
    # Obtener embeddings para todos los textos
    vectores = embeddings.embed_documents(textos)
    
    print(f"Número de vectores generados: {len(vectores)}")
    print(f"Dimensión de cada vector: {len(vectores[0])}")
    
except Exception as e:
    print(f"No se pudo generar embeddings con LangChain: {e}")
    print("Para usar esta funcionalidad, necesitas configurar tu API key.")
```
